```python
# Q0
for x in np.linspace(1,7,500):
    y=x**4*exp(-2*x)
    z=(x**2*exp(-x)*sin(x**2))**2
    plt.plot(x,y,'go-')
    plt.plot(x,z,'r^')
plt.xlabel('X axis')
plt.ylabel('Y(green) Z(red) axis')
plt.title('XYZ graph')
#plt.legend(loc='upper right')
```

```python
#Q1
data=np.genfromtxt('microphone.txt')
Freq=data[:,1]
M1=data[:,2]
M2=data[:,3]
Ratio=M1/M2
plt.plot(Freq,Ratio,'r-',label='ratio')
plt.xlabel('Frequency')
plt.ylabel('Ratio of M1 M2')
plt.title('Freq vs Signal ratio')
plt.legend(loc='upper left')
```
