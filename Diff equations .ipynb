```python
# Solving an ODE
import numpy as np
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def g(y, x):
    y0 = y[0]
    y1 = y[1]
    y2 = ((3*x+2)*y1 + (6*x-8)*y0)/(3*x-1)
    return y1, y2

# Initial conditions on y, y' at x=0
init = 2.0, 3.0
# First integrate from 0 to 2
x = np.linspace(0,2,100)
sol=odeint(g, init, x)
plt.plot(x, sol[:,0], color='b')
# Then integrate from 0 to -2
x = np.linspace(0,-2,100)
sol=odeint(g, init, x)
plt.plot(x, sol[:,0], color='b')

# The analytical answer in red dots
exact_x = np.linspace(-2,2,10)
exact_y = 2*np.exp(2*exact_x)-exact_x*np.exp(-exact_x)
plt.plot(exact_x,exact_y, 'o', color='r', label='exact')
plt.legend()

plt.show()
```

```python
# 2nd order ODE
# Y"+2Y'+2Y=cos(2x); y(0)=0,Y'(0)=0
# Let Z=Y'
# Z'+2Z+2Y=cos(2x); Z(0)=0,Y(0)=0
```

```python
def dU_dx(U,x):
    return[U[1],-2*U[1]-2*U[0]+np.cos(2*x)]
U0=[0,0]
x=np.linspace(0,10,50)
U=odeint(dU_dx,U0,x)
y=U[:,1]
plt.plot(x,y)
```

```python
#Solve 1st order ODE, Ideal simple pendulum
# theta"=-(g/l)sin(theta)
# Let theta'=Y
# Y'=-(g/l)sin(theta)
```

```python
g=9.8;l=1
def dth_dt(theta,t):
    return -(g/l)*sin(t)
t=np.linspace(0,40,150)
Sol=odeint(dth_dt,10,t)
plt.plot(t,Sol)
```

```python
# Real pendulum
# theta"=-(g/l)sin(theta)-b*theta'
```

```python
g=9.8;l=1;b=0.2
def dth_dt(theta,t):
    return -(g/l)*sin(t)-b*t
t=np.linspace(0,40,150)
Sol=odeint(dth_dt,10,t)
plt.plot(t,Sol)
```

```python
# Solve 2nd order ODE
# theta'(t) = omega(t)
# omega' = -b*omega - (g/l)*sin(theta)
```

```python
def pend(y, t):
    theta, omega = y
    return [omega, -b*omega - (g/l)*np.sin(theta)]

b = 0.5;g = 9.8;l = 1
y0 = [np.pi - 0.1, 0.0]
t = np.linspace(0, 10, 101)

sol = odeint(pend, y0, t)

plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
# from https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.odeint.html
```

```python
def dudx(u,x):
    return [u[1],-2*u[1]-2*u[0]+np.cos(x**2)]
ui=[0,0]
x=np.linspace(0,10,100)
sol=odeint(dudx,ui,x)
plt.plot(x,sol[:,0])
```

```python
# Driving damped oscillator
a=1.5;b=0.5;g=10;l=1;w=0.6
def func(th,t):
    return [th[1],-g/l*sin(th[0])-b*th[1]+a*cos(w*t)]
thi=[0,0]
t=np.linspace(0,70,100)
sol=odeint(func,thi,t)
plt.plot(t,sol[:,0])
```

```python
#Phase space plots .
plt.plot(sol[:,1],sol[:,0])
```

```python
#Duffing Oscillator
a=1;b=-1;d=0.1;g=0.3;w=1
def eqn(x,t):
    return [x[1],-d*x[1]-b*x[0]-a*x[0]**3+g*cos(w*t)]
x=[0,0]
t=np.linspace(0,50,100)
sol=odeint(eqn,x,t)
plt.plot(t,sol[:,1])
```

```python
#Phase space plots
plt.plot(sol[0],sol[1])
```

```python
# Integrating using traditional method, grid based method.

nrectangles = 100 # number of x-values to use to calculate the average.

# Integration domain
xlow = 0.
xhigh = 3.

# Integrand
f = lambda x: sin(x)

xvalues = np.linspace(xlow, xhigh, nrectangles)
fvalues = f(xvalues) # f(x_i) for each rectangle
areas = fvalues * (xhigh - xlow)/nrectangles # A_i for each rectangle
integral = sum(areas)

print("Integral: ", integral)
```
